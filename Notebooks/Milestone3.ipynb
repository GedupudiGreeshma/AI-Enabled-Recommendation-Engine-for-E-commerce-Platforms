```python
import pandas as pd
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# Load user-item matrix
ui_matrix = pd.read_csv("user_item_matrix.csv", index_col=0)

ui_matrix.head()
```

```python
train_matrix = ui_matrix.copy()
test_data = {}

for user in ui_matrix.index:
    interacted_items = ui_matrix.loc[user][ui_matrix.loc[user] > 0].index.tolist()

    if len(interacted_items) > 0:
        test_item = np.random.choice(interacted_items)
        train_matrix.loc[user, test_item] = 0
        test_data[user] = test_item
```

```python
# Item popularity from training data
item_popularity = train_matrix.sum(axis=0)

def recommend_items(user, top_k=5):
    user_interactions = train_matrix.loc[user]
    unseen_items = user_interactions[user_interactions == 0].index
    ranked_items = item_popularity[unseen_items].sort_values(ascending=False)
    return ranked_items.head(top_k).index.tolist()
```

```python
y_true = []
y_pred = []

TOP_K = 5

for user, actual_item in test_data.items():
    recommendations = recommend_items(user, TOP_K)

    y_true.append(1)
    y_pred.append(1 if actual_item in recommendations else 0)
```

```python
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print("Model Evaluation Metrics:")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
```

```python
for k in [3, 5, 10]:
    y_pred_k = []

    for user, actual_item in test_data.items():
        recs = recommend_items(user, top_k=k)
        y_pred_k.append(1 if actual_item in recs else 0)

    print(f"\nTop-{k} Recommendations")
    print("Precision:", precision_score(y_true, y_pred_k))
    print("Recall:", recall_score(y_true, y_pred_k))
    print("F1-score:", f1_score(y_true, y_pred_k))
```
